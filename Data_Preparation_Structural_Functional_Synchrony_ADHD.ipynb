```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
''' First of all Taking the specific Phenotype Information and combining it 
with AI information'''
# Reading the Cintya's file on Structural information

structural_data_path = '/content/ADHD_200_FunStr_Cintya.xlsx'
structural_data_all = pd.read_excel(structural_data_path)
```

```python
structural_data_all.head()
```

```python
# Number of Unique Participants in the study
print('Number of Unique Participants in the study : {}'.format(len(structural_data_all['ScanDir.ID'].unique())))
```

```python
# Looking at the repeated Measurements
structural_data_all['ScanDir.ID'].value_counts().unique()
```

```python
# Saving the Regions Mapping 
regions_mapping = {}
for key,val  in zip(structural_data_all['RG'],structural_data_all['Region']):
  if key not in regions_mapping.keys():
    regions_mapping[key] = val

print('We obtained a mapping with {} Regions'.format(len(regions_mapping))) 
```

```python
# Now converting the data in the wide format
import time
#k=structural_data_all[structural_data_all['ScanDir.ID'] == scan_id]['RG']
#v=structural_data_all[structural_data_all['ScanDir.ID'] == scan_id]['AI']
start = time.time()
participants_structural_data = {}
for scan_id in structural_data_all['ScanDir.ID']:
  regions = structural_data_all[structural_data_all['ScanDir.ID'] == scan_id]['RG']
  ai_vals = structural_data_all[structural_data_all['ScanDir.ID'] == scan_id]['AI']
  reg_ai = {r:v for r,v in zip(regions,ai_vals)}
  participants_structural_data[scan_id] = reg_ai
end = time.time()
print('Total Time Taken --> : {} seconds'.format(end-start))
```

```python
long_format_structural_data_all_participants = pd.DataFrame(participants_structural_data).T
```

```python
long_format_structural_data_all_participants=long_format_structural_data_all_participants.reset_index()
```
